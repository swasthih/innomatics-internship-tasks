# Problem 1: Smart Parking Lot Management System

```python
def smart_parking_system(capacity, vehicle_logs):
    """
    Manages parking lot entries and exists.
    Calculates current parked vehicles and checks capacity.
    """

    parked = 0
    peak_usage = 0

    #Process vehicle logs
    for log in vehicle_logs:
        if log == "IN":
           parked += 1
        elif log == "OUT" and parked > 0:
           parked -= 1

        #Track peak usage
        if parked > peak_usage:
            peak_usage = parked

    #Parking status check
    if parked > capacity:
        status > "Over Capacity"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", parked)
    print("Peak Parking Usage:", peak_usage)
    print("Parking Status:", status)

#Example 
smart_parking_system(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])         
```

# Problem 2: Online Food Delivery Time Estimator 

```python
def estimate_delivery_time(distance, traffic, weather):
    """
    Estimates delivery time based on distanace, traffic and weather.
    """

    # Base time calculation (5 minutes per km)
    time = distance * 5

    # Traffic delay
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10
    elif traffic == "Low":
        time += 5

    #weather delay
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    print("Estimated Delivery Time:", time, "minutes")

#Example
estimate_delivery_time(8, "High", "Rainy")
```

# Problem 3: Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy(total_seats, booked_seats):
    """
    Calculates occupancy percentage and determines show status
    """

    booked_count = len(booked_seats) 
    occupancy = (booked_count / total_seats) * 100

    #Show classification
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Seats available"

    print("Occupancy:", round(occupancy), "%")
    print("Show Status:", status)

#Example (150 booked seats out of 200)
seat_occupancy(200, [1]*150)
```

# Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    """
    Calculates average CPU load and classifies server 
    """

    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else: status = "Critical"

    print("Average CPU Load:", round(average), "%")
    print("Server Status:", status)

#Example 
classify_server_load([45, 60, 70, 85, 90])
        
```
